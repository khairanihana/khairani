```python
mahasiswa = {
    "10121001": "Asep",
    "10121002": "Budi",
    "10121003": "Cecep"
}

nilai = {
    "10121001": [50, 70, 40, 80],
    "10121002": [78, 78, 80, 65],
    "10121003": [57, 88, 67, 69]
}

tertinggi = 0
nim_terbaik = ""

for nim in nilai:
    total = 0
    for n in nilai[nim]:
        total = total + n

    rata = total / len(nilai[nim])

    if rata > tertinggi:
        tertinggi = rata
        nim_terbaik = nim

print("Mahasiswa Terpintar :", mahasiswa[nim_terbaik], "Nilai:", round(tertinggi, 2))

jumlah_mk = [0, 0, 0, 0]

for nim in nilai:
    for i in range(len(nilai[nim])):
        jumlah_mk[i] = jumlah_mk[i] + nilai[nim][i]

terkecil = 100
mk_ke = 0

for i in range(len(jumlah_mk)):
    rata = jumlah_mk[i] / len(nilai)

    if rata < terkecil:
        terkecil = rata
        mk_ke = i + 1

print("Mata Kuliah Nilai Terkecil : MK", mk_ke, "Nilai:", round(terkecil, 2))
```

```python
while True:
    print()
    print("=== MENU ===")
    print("1. Penjumlahan")
    print("2. Pengurangan")
    print("3. Perkalian")
    print("0. Exit")

    pilih = int(input("Pilih menu: "))

    if pilih == 0:
        print("Program selesai")
        break

    baris = int(input("Masukkan jumlah baris: "))
    kolom = int(input("Masukkan jumlah kolom: "))

    print("\nInput Matriks A")
    A = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            angka = int(input("A[" + str(i) + "][" + str(j) + "] = "))
            row.append(angka)
        A.append(row)

    print("\nInput Matriks B")
    B = []
    for i in range(baris):
        row = []
        for j in range(kolom):
            angka = int(input("B["+ str(i) + "][" + str(j) + "] = "))
            row.append(angka)
        B.append(row)

    hasil = []

    if pilih == 1:
        for i in range(baris):
            row = []
            for j in range(kolom):
                row.append(A[i][j] + B[i][j])
            hasil.append(row)

    elif pilih == 2:
        for i in range(baris):
            row = []
            for j in range(kolom):
                row.append(A[i][j] - B[i][j])
            hasil.append(row)

    elif pilih == 3:
        if kolom != baris:
            print("Perkalian tidak bisa! ukuran tidak sesuai")
            continue

        for i in range(baris):
            row = []
            for j in range(kolom):
                total = 0
                for k in range(kolom):
                    total += A[i][k] * B[k][j]
                row.append(total)
            hasil.append(row)

    else:
        print("Pilihan tidak ada")
        continue

    print()
    print("Hasil:")
    for i in hasil:
        print(i)
```
