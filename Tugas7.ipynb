```python
while True:
    print("\n=== MENU ===")
    print("1. Pangkat")
    print("2. Hitung 1 - 2/3 + 5/8 - 13/21 + ...")
    print("0. Exit")

    f = input("Pilih menu: ")

    if f == "1":
        bilangan = int(input("Masukkan bilangan: "))
        pangkat = int(input("Masukkan pangkat: "))

        hasil = 1
        for i in range(pangkat):
            hasil *= bilangan

        print("===================================")
        print("Hasil dari", bilangan, "^", pangkat, "=", hasil)
        print("===================================")

    elif f == "2":
        n = int(input("Masukkan jumlah suku: "))
        hasil = 0
        a, b = 1, 2
        c, d = 1, 3

        deret = ""

        for i in range(n):
            if i == 0:
                deret = deret + str(a) + "/" + str(c)
                hasil += a / c
            else:
                if i % 2 == 0:
                    deret = deret + " + " + str(a) + "/" + str(c)
                    hasil += a / c
                else:
                    deret = deret + " - " + str(a) + "/" + str(c)
                    hasil -= a / c

            a, b = b, a + b
            c, d = d, c + d

        print("Deret:", deret)
        print("===================================")
        print("Hasil dari", n, "suku =", hasil)
        print("===================================")

    elif f == "0":
        print("Program selesai.")
        break

    else:
        print("Pilihan tidak tersedia!")
```
